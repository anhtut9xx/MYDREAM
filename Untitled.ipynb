```python
# Bài tập Câu Điều Kiện:

#[Bài tập 1] Kiểm tra số nhập vào là số chẳn hoặc lẻ, sử dụng toán tử điều kiện ba ngôi:

num = int(input("Nhập vào một số nguyên: "))
result = "số chẵn" if num % 2 == 0 else "số lẻ"
print(f"{num} là {result}")


#[Bài tập 2] Kiểm tra số chẵn/lẻ:

a = input("Nhập số a là: ")
a = float(a)
if a % 2 == 0:
    print('số chẵn')
elif a % 2 == 1:
    print('số lẻ')
else:
    print('không phải số tự nhiên')

#[Bài tập 3] Hiển thị ngày trong tuần:

weekdays = int(input("Give me a number (form 1 to 7): "))
if weekdays == 1:
    print("Monday")
elif weekdays == 2:
    print("Tuesday")
elif weekdays == 3:
    print("Wednesday")
elif weekdays == 4:
    print("Thursday")
elif weekdays == 5:
    print("Friday")
elif weekdays == 6:
    print("Saturday")
elif weekdays == 7:
    print("Sunday")
else:    
    print("Error, out of range")
    
#[Bài tập 4] Hiển thị thông tin đơn hàng:

amount = float(input("Hãy nhập số tiền bạn đã chi: "))

if amount < 75:
    discount = 0
elif amount < 100:
    discount = 15
elif amount < 150:
    discount = 25
else:
    discount = 50

total_amount = amount - discount
print(f"Số tiền bạn phải thanh toán là: {total_amount}$")

#[Bài tập 5] Tính chỉ số cân nặng của cơ thể:

weight = int(input("Nhập số cân nặng của bạn: "))
height = float(input("Nhập số chiều cao của bạn: "))

BMI = weight / (height ** 2)

if BMI < 16:
    print("Gầy cấp độ III")
elif 16 <= BMI < 17:
    print("Gầy cấp độ II")
elif 17 <= BMI < 18.5:
    print("Gầy cấp độ I")
elif 18.5 <= BMI < 25:
    print("Bình thường")
elif 25 <= BMI < 30:
    print("Thừa cân")
elif  30 <= BMI < 35:
    print("Béo phì cấp độ I")
elif 35 <= BMI < 40:
    print("Béo phì cấp độ II")
else:
    print("Béo phì cấp độ III")
    
import tkinter as tk

def btn_click(btn_val):
    current = input_box.get()
    new = current + str(btn_val)
    input_box.delete(0, tk.END)
    input_box.insert(0, new)

def clear():
    input_box.delete(0, tk.END)

def calculate():
    try:
        expression = input_box.get()
        result = eval(expression)
        input_box.delete(0, tk.END)
        input_box.insert(0, str(result))
    except:
        input_box.delete(0, tk.END)
        input_box.insert(0, "Lỗi")
```

```python
#[Bài tập] Simple GUI calculator using Tkinter:
# Đề bài: Viết chương trình tạo ứng dụng máy tính đơn giản cho người dùng, thực hiện các phép tính +, -, *, / .Có thể nhập được số lẻ và có chức năng clear.

import tkinter as tk

root = tk.Tk()
root.title("Máy Tính Đơn Giản")

# Tạo ô nhập liệu
input_box = tk.Entry(root, font=("Helvetica", 16), justify=tk.RIGHT)
input_box.grid(row=0, column=0, columnspan=4)

# Tạo các nút số
for i in range(9):
    btn = tk.Button(root, text=str(i + 1), font=("Helvetica", 16), command=lambda i=i: btn_click(i + 1))
    btn.grid(row=1 + i // 3, column=i % 3)

# Tạo các nút phép tính
operators = ['+', '-', '*', '/', '.']
for i, operator in enumerate(operators):
    btn = tk.Button(root, text=operator, font=("Helvetica", 16), command=lambda operator=operator: btn_click(operator))
    btn.grid(row=i + 1, column=3)

# Tạo nút xóa
clear_btn = tk.Button(root, text='C', font=("Helvetica", 16), command=clear)
clear_btn .grid(row=5, column=0)

# Tạo nút kết quả
equal_btn = tk.Button(root, text='=', font=("Helvetica", 16), command=calculate)
equal_btn.grid(row=5, column=1)

# Chạy vòng lặp giao diện đồ họa
root.mainloop()
```
